```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

DATA

```python
saude = pd.read_csv('Deepression.csv')
saude.head()
```

```python
# eliminando a coluna Number pois ela não influenciara nos  resultados
if 'Number ' in saude.columns:
    saude.drop(columns=['Number '], inplace=True)
elif 'Number' in saude.columns:
    saude.drop(columns=['Number'], inplace=True)
else:
    print("Coluna 'Number' não encontrada.")
```

```python
#mostrando as colunas sem a coluna Number
print(saude.head())
```

```python
# checando se há valores nulos
saude.isnull().any()
```

```python
#Verificar quais colunas estão vazias
saude.columns[saude.isnull().all()]
```

```python
#Tratando as colunas
saude.dropna(axis=1, how='all', inplace=True)
```

```python
saude.dropna(axis=0, how='any', inplace=True)
```

```python
# verifica valores nulos ou ausentes
saude.isnull().sum()
```

```python
saude.head()
```

```python
nomes_colunas = ['Fatigue', 'Worthlessness', 'Agitation', 'Suicidal Ideation', 
                 'Aggression', 'Panic Attacks', 'Restlessness', 'Low Energy']
```

```python
# Sintomas mais comuns de paciente em estado de depressão
freq_sintomas = saude[nomes_colunas].apply(pd.Series.value_counts)
freq_sintomas
```

Sintomas Mais Comuns:

Observando as contagens mais altas em cada coluna, podemos identificar quais sintomas são mais frequentemente relatados pelos pacientes.
Por exemplo, se 'Fatigue' tem a maioria das contagens em nível 5, isso sugere que fadiga intensa é um sintoma comum entre os pacientes em estado de depressão.

```python
# monstrando isso em gráfico
plt.figure(figsize=(10, 6))
freq_sintomas.plot(kind='bar', ax=plt.gca())
plt.title('Frequência dos Sintomas')
plt.xlabel('Sintomas')
plt.ylabel('Frequência')
plt.xticks(rotation=45)
plt.legend(title='Sintomas')
plt.tight_layout()
plt.show()
```

```python
# Calcular a gravidade como a média dos sintomas
saude['GRAVIDADE'] = saude[nomes_colunas].mean(axis=1)
```

```python
saude.head()
```

```python
sns.heatmap(saude.isnull())
```

```python
import statsmodels.api as sm 

sns.displot(saude['GRAVIDADE'], kde=True)
```

É importante observar que nem todas as pessoas com depressão apresentam todos esses sintomas, e a gravidade 
e a combinação de sintomas podem variar amplamente entre os indivíduos.Os sintomas como agitação, fadiga,

```python
# modelo para prever os sintomas mais mencionados
x = saude[['Fatigue', 'Worthlessness', 'Agitation', 'Suicidal Ideation', 
            'Aggression', 'Panic Attacks', 'Restlessness', 'Low Energy']]
y = saude['GRAVIDADE']
```

```python
from sklearn import preprocessing, svm
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
# dividir conjunto de treino e teste

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=35)
```

```python
LinearRegression()
```

```python
# Treinar o modelo
lr.fit(x_train, y_train)
```

```python
# Prever os valores no conjunto de teste
y_pred = lr.predict(x_test)
```

```python
from sklearn.metrics import r2_score
```

```python
# Calcular o r2_score para avaliar o modelo
r2 = r2_score(y_test, y_pred)
print(f'R² Score: {r2}')
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators=100, random_state=35)
regressor.fit(x_train, y_train)

y_pred = regressor.predict(x_test)
r2_score(y_test, y_pred)
print("Accuracy score (training): {0:.3f}".format(regressor.score(x_train, y_train)))
print("Accuracy score (predicted): {0:.3f}".format(regressor.score(x_test, y_test)))
```

```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error as MSE
from sklearn import datasets
  

gbr = GradientBoostingRegressor(n_estimators = 200, max_depth = 1)
  
# Fit to training set
gbr.fit(x_train, y_train)
  
# Predict on test set
y_pred = gbr.predict(x_test)
  
# test set RMSE
test_rmse = MSE(y_test, y_pred) ** (1 / 2)
  
# Print rmse
print('RMSE test set: {:.2f}'.format(test_rmse))
r2_score(y_test, y_pred)
```

```python
import matplotlib.pyplot as plt

# Criar um gráfico de dispersão para comparar as previsões com os valores reais
plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_pred, color='blue')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=2)
plt.xlabel('Valores Reais da Gravidade dos Sintomas')
plt.ylabel('Previsões da Gravidade dos Sintomas')
plt.title('Comparação entre Previsões e Valores Reais da Gravidade dos Sintomas')
plt.show()
```
